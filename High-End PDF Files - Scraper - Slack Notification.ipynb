```python
def post_message_to_slack(slack_token, text, channel, blocks=None):
    try:
        print('Sending message...')
        client = slack_sdk.WebClient(token=SLACK_TOKEN)
        request = client.chat_postMessage(text=text,channel=channel,blocks=blocks,icon_url=ICON,username=USERNAME)
        print('Message was sent successfully.')
        return request
    except Exception as e:
        print(f'Error: {e}')
```

```python
def base64_to_pdf(base64String, idx):
    data = base64.b64decode(base64String)
    file_name = f"{folder_dir}/{idx}.pdf"
    with open(file_name, "wb") as f:
        f.write(data)
    return file_name
```

```python
def get_data(url, idx):
    payload['params']['urls'] = [url]
    try:
        response = http.post(URL, json=payload, headers=headers)
    except requests.ConnectionError as e:
        print(e[:55])
    if response.status_code == 200:
        resp_result = response.json()
        final_result = resp_result['result'].split('data:application/pdf;base64,')[-1]
        base64_to_pdf(final_result, idx)
    else:
        errorsUrls.append(url)
```
